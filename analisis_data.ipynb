```python
import pandas as pd
```

```python
data_guru = pd.read_csv('datasets/processed-data-guru-kelompok-usia-provinsi-2012.csv', delimiter=",", encoding="ISO-8859-1")

data_guru
```

## 1. Deskripsi Data

Data merupakan data guru menurut kelompok usia pada tahun 2012 yang disimpan berdasarkan kolom:
- kode_provinsi
- nama_provinsi
- tingkat_pendidikan
- kelompok_usia
- jumlah_guru

Terdapat sebanyak 1386 baris data pada semua kolom kecuali kolom jumlah guru yang memiliki 1374 baris data

## 2. Statistik tentang data

```python
# jumlah guru keseluruhan

data_guru.jumlah_guru.sum()
```

```python
# median pada kolom jumlah guru

data_guru.jumlah_guru.median()
```

```python
# standar deviasi pada kolom jumlah guru

data_guru.jumlah_guru.std()
```

```python
# rata-rata jumlah guru menurut tingkat pendidikan

data_guru.groupby('tingkat_pendidikan').sum()['jumlah_guru']
```

```python
# rata-rata jumlah guru menurut kelompok usia

data_guru.groupby('kelompok_usia').sum()['jumlah_guru']
```

```python
# jumlah guru berdasarkan nama provinsi dan

data_guru.groupby(['tingkat_pendidikan','kelompok_usia']).sum()['jumlah_guru']
```

## 3. Visualisasi

```python
ax = data_guru.groupby(['tingkat_pendidikan','kelompok_usia']).sum()['jumlah_guru'].T
ax = ax.plot(kind='bar', title="Rata-Rata Jumlah Guru Berdasarkan Kelompok Usia", legend=True, fontsize=14, figsize=(18,6))
ax.set_xlabel('Kelompok Usia Menurut Tingkat Pendidikan', fontsize=14)
ax.set_ylabel('Jumlah Guru', fontsize=14)
```

```python
ax = data_guru.groupby('kelompok_usia').sum()['jumlah_guru']
ax = ax.plot(kind='bar', title="Rata-Rata Jumlah Guru Berdasarkan Kelompok Usia", legend=True, fontsize=12)
ax.set_xlabel('Kelompok Usia', fontsize=14)
ax.set_ylabel('Jumlah Guru', fontsize=14)
```

```python
ax = data_guru.groupby('tingkat_pendidikan').sum()['jumlah_guru']
ax = ax.plot(kind='bar', title='Rata-Rata Jumlah Guru Berdasarkan Tingkat Pendidkan', legend=True, fontsize=12)
ax.set_xlabel('Tingkat Pendidikan', fontsize=14)
ax.set_ylabel('Jumlah Guru', fontsize=14)
```
