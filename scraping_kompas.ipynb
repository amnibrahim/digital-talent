```python
import urllib
from bs4 import BeautifulSoup
```

```python
# get html elements

r = urllib.urlopen('http://tekno.kompas.com').read()

print(type(r))
print(len(r))
print(r[:1000])
```

```python
soup = BeautifulSoup(r,"lxml")

print(type(soup))
print(soup.prettify()[:1000])
```

```python
# get all <a/> links

links = soup.find_all('a')
print(len(links))

for i in links:
    print("%s: %s" % (i.text.strip(),i['href']))
```

```python
# get all berita links

link_berita = [i for i in links if 'tekno.kompas.com/read/' in str(i) and 'javascript:void' not in str(i) and '#' not in str(i)]

print(len(link_berita))

for i in link_berita:
    print("%s: %s" % (i.text.strip(),i['href']))
```

```python
# get judul, isi, date from 1 berita page

html = urllib.urlopen('https://tekno.kompas.com/read/2018/10/25/20070037/ini-spesifikasi-dan-harga-xiaomi-mi-mix-3').read()

soup = BeautifulSoup(html, 'lxml')

judul = soup.find('h1','read__title').text.strip()
isi = soup.find('div','read__content').text.strip()
date = soup.find('div','read__time').text.strip().split('-')[1].strip().split(',')[0]

print(judul)
print(date)
print(isi)
```

```python
# get judul, isi, date from all berita pages

def getData(link):
    html = urllib.urlopen(link).read()
    soup = BeautifulSoup(html, 'lxml')

    judul = soup.find('h1','read__title').text.strip()
    isi = soup.find('div','read__content').text.strip()
    date = soup.find('div','read__time').text.strip().split('-')[1].strip().split(',')[0]
    
    return [judul, isi, date]
```

```python
dataset_berita = []

for i in link_berita:
    print('%s: %s' % (i.text.strip(),i['href']))
    dataset_berita.append(getData(i['href']))
```
